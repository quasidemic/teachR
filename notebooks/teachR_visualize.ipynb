# Visualization

- `ggplot2`

Cheatsheet: https://raw.githubusercontent.com/rstudio/cheatsheets/master/data-visualization-2.1.pdf

```r
library(readr)

ess2014_p1 <- read_csv("https://github.com/CALDISS-AAU/workshop_r-table-data/raw/master/data/ess2014_mainsub_p1.csv")
```

```r
library(ggplot2)

ggplot(ess2014_p1, aes(x = yrbrn, y = weight, colour = gndr)) + 
    geom_point()
```

```r
ggplot(ess2014_p1, aes(x = yrbrn, y = weight, colour = gndr)) + 
    geom_point() + 
    geom_smooth()
```

```r
ggplot(ess2014_p1, aes(x = yrbrn, y = weight, colour = gndr)) + 
    geom_point() + 
    geom_smooth(method = "lm")
```

```r
ggplot(ess2014_p1, aes(x = gndr, y = weight, colour = gndr)) + 
    geom_jitter()
```

```r
ggplot(ess2014_p1, aes(x = height)) + 
    geom_histogram()
```

```r
ggplot(ess2014_p1, aes(x = gndr, fill = cgtsmke)) + 
    geom_bar()
```

```r
ggplot(ess2014_p1, aes(x = gndr, y = weight, fill = cgtsmke)) + 
    geom_boxplot()
```
